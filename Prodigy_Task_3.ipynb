```python
high_corr_cols=['emp.var.rate','euribor3m','nr.employed']
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df_encoded=df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
x=df_encoded.drop('deposit',axis=1)  #independent variable
y=df_encoded['deposit']   #dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def eval_model(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy Score:", acc)
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Function to evaluate model performance
def eval_model(y_test, y_pred):
    # Calculate accuracy
    acc = accuracy_score(y_test, y_pred)
    
    # Display results
    print("Accuracy Score:", acc)
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))


```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=10)

# Fit the model using the training data
dt.fit(x_train, y_train)
```

```python
# Get the training score (accuracy on training data)
training_score = dt.score(x_train, y_train)
print("Training Score:", training_score)

# Get the testing score (accuracy on test data)
testing_score = dt.score(x_test, y_test)
print("Testing Score:", testing_score)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Predict the labels for the test set
y_pred = dt.predict(x_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.metrics import classification_report

# Generate the classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```
